```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
def plotting(data, string, title):
    # Extracting data ranges
    multipole_range = [list(map(int, d[1].split("-"))) for d in data]
    frequency_range = [list(map(int, d[2].split("-"))) for d in data]

    # Plotting rectangles
    fig, ax = plt.subplots(figsize=(7, 4))
    patches = []

    colors = plt.cm.tab20(range(len(data)), alpha=0.5)
    
    for i in range(len(data)):
        rect = plt.Rectangle((multipole_range[i][0], frequency_range[i][0]),
                             multipole_range[i][1] - multipole_range[i][0],
                             frequency_range[i][1] - frequency_range[i][0],
                             edgecolor=colors[i], linewidth=3, linestyle = 'dotted')
        ax.add_patch(rect)
        patches.append(rect)

    # Set colors with transparency for each rectangle
    for rect, color in zip(patches, colors):
        rect.set_facecolor(color)

    # Creating legend with modified names
    legend_labels = [name[0] for name in data]
    legend_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colors, legend_labels)]
    plt.legend(handles=legend_patches, loc='center left', bbox_to_anchor=(1, 0.5), title="Experimentos")

    # Set plot labels and title
    plt.xlabel('Rango del multipolo ($\mathcal{l}$)')
    plt.ylabel('Rango de frecuencia (GHz)')
    plt.title(title)

    ax.set_xscale('log')
    ax.set_yscale('log')

    # Display the plot
    plt.grid(True, zorder=0)  # Display gridlines behind the rectangles
    plt.tight_layout()
    
    plt.savefig(string, dpi=300, bbox_inches='tight')
    plt.show()
```
